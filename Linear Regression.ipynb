```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"Dataset/SalaryData.csv", encoding='latin-1')
```

```python
X=data['YearsExperience'].values
Y=data['Salary'].values
```

```python
covariance = np.cov(X, Y)[0, 1]
```

```python
variance=np.var(X)
```

```python
slope=covariance/variance
```

```python
intercept=np.mean(Y)-(np.mean(X)*slope)
```

```python
plt.scatter(X,Y, color='blue', label='Data points')
plt.plot(X,slope * X + intercept,color='red', label='Regression line')
plt.xlabel("YearsExperience")
plt.ylabel("Salary")
plt.legend()
plt.show()
```

```python
new_exp=8.9
prediction=slope*new_exp +intercept
print(f"A person with {new_exp} years can have a salary of {int(prediction)}")
```
